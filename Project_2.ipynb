```python
X=file.drop(["BodyFat","Age"],axis=1)
X.head()
#X.tail()
```

```python
Y=file["BodyFat"]
Y.head()
#Y.tail()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# X_train
X_train.head()
# X_train.tail()
```

```python
# X_test
X_test.head()
# X_test.tail()
```

```python
Y_train
Y_train.head()
# Y_train.tail()
```

```python
Y_test
Y_test.head()
# Y_test.tail()
```

```python
from sklearn.linear_model import LinearRegression
Model=LinearRegression()
Model.fit(X_train,Y_train)
```

```python
Y_pred = Model.predict(X_test)
Y_pred
# Y_test
```

```python
# from sklearn.metrics import accuracy_score
# Accuracy = accuracy_score(Y_test,Y_pred)
# Accuracy
```

```python
# Plotting the actual vs predicted BodyFat
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, Y_pred, color='blue', alpha=0.5)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='red', linewidth=2)
plt.title('Actual vs Predicted Body Fat Percentage')
plt.xlabel('Actual Body Fat')
plt.ylabel('Predicted Body Fat')
plt.grid(True)
plt.show()
```

```python
import joblib
joblib.dump(Model,"Body_Fat.pkl")
```
